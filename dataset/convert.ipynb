```python
with open(r'./pattern.txt') as f:
    l = f.readline()
    g={}
    node={}
    size =0 
    while l:

        s, t = list(map(int, l.strip("\n").split(" ")))
        if g.get(s) == None: g[s]={}
        if g[s].get(t) == None: g[s][t] = 1

        if node.get(s) == None: 
            node[s]=1
            size += 1
        if node.get(t) == None:
            node[t] = 1
            size += 1

        l = f.readline()

output = ""
for i in range(size):
    for j in range(size):
        if g.get(i) != None and g[i].get(j) != None:
            output += "1"
        else: output += "0"

print(f"{size} {output}")
```

```python
split=[]
path=[]

spnode={}
pnode={}

with open(r"./splitPath.txt") as f:
    line = f.readline()
    while line:

        line = list(map(int,line.strip("\n").split()))

        u, v = line

        split.append((u,v))

        if spnode.get(u) == None: spnode[u] = 1
        if spnode.get(v) == None: spnode[v] = 1

        line = f.readline()

with open(r"./treePath.txt") as f:
    line = f.readline()
    while line:

        line = list(map(int,line.strip("\n").split()))

        u, v = line

        path.append((u,v))

        if pnode.get(u) == None: spnode[u] = 1
        if pnode.get(v) == None: spnode[v] = 1

        line = f.readline()

print("error match:")
for i in path:
    if i not in split:
        print(i)

print("lack:")
for i in split:
    if i not in path:
        print(i)

print("error match node:")
for i in pnode.keys():
    if(spnode.get(i) == 0): print(i)
```

```python
brute_force={}
with open(r'./brute_force.log') as f:
    line = f.readline()
    while line:

        pair = list(map(int, line.strip(" \n").split(" ")))
        # pair = tuple(sorted(pair))
        pair = tuple(pair)
        if brute_force.get(pair) == None:
            brute_force[pair] = 1
        else:
            print("duplicate embedding in brute_force")
            assert(0)
        
        line = f.readline()

temporal={}
with open(r'./temporal.log') as f:
    line = f.readline()
    while line:

        pair = list(map(int, line.strip(" \n").split(" ")))
        # pair = tuple(sorted(pair))
        pair = tuple(pair)
        if temporal.get(pair) == None:
            temporal[pair] = 1
        else:
            print("duplicate embedding in temporal")
            assert(0)
        
        line = f.readline()

for i in temporal.keys():
    if i not in brute_force.keys():
        print(i)

print("lack: ")
for i in brute_force.keys():
    if i not in temporal.keys():
        print(i)

```
